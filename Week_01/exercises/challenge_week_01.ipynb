# Import Required Libraries
Import the necessary libraries, including NumPy, Matplotlib, and system information libraries.

```python
# Import Required Libraries

import numpy as np
import matplotlib.pyplot as plt
import os
import platform
import socket
from platform import python_version
from datetime import datetime
```

# Create Numpy Array with Random Numbers
Use NumPy to create an array with random numbers.

```python
# Create Numpy Array with Random Numbers

# Create a numpy array with random numbers
random_numbers = np.random.rand(100)  # Generate an array of 100 random numbers between 0 and 1
```

# Calculate Quartiles
Calculate the quartiles of the random numbers using the percentile() method from the NumPy library.

```python
# Calculate Quartiles

# Calculate the quartiles of the random numbers using the percentile() method from the NumPy library
quartiles = np.percentile(random_numbers, [25, 50, 75])
quartiles
```

# Generate Histogram
Create a histogram based on the numpy array using Matplotlib.

```python
# Generate Histogram

# Create a histogram based on the numpy array using Matplotlib
plt.hist(random_numbers, bins=10, edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Generate Boxplot
Create a boxplot based on the numpy array using Matplotlib.

```python
# Generate Boxplot

# Create a boxplot based on the numpy array using Matplotlib
plt.boxplot(random_numbers)
plt.title('Boxplot of Random Numbers')
plt.ylabel('Value')
plt.show()
```
